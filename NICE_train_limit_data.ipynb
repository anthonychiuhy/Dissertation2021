```python
import torch

from flows import ObjectDataset, MLPLayers, AdditiveCouplingLayers, Flow
from utils import save_object, load_object

from Datasets import Spiral, Funnel, Ring, Grid, Banana, Uniform, Cosine, Multiring
```

```python
# Spiral, Funnel, Ring, Grid, Banana, Uniform, Cosine, Multiring
d = 1
D = 2
layers = 20

sample_size = 5000
batch_size = 100
epochs = 200
lr = 0.001
weight_decay = 0.01

ensembles = 20
save_path = './saved_flows_additive_limit_data/'


names = ['Funnel', 'Banana', 'Ring', 'Uniform', 'Cosine', 'Grid', 'Multiring']
objs = [Funnel(), Banana(), Ring(), Uniform(), Cosine(), Grid(), Multiring()]

for i in range(len(objs)):
    name = names[i]
    obj = objs[i]
    
    for j in range(30, 30+ensembles):
        while True:
            try:
                flow = Flow(d, D, layers, coupling_layers='additive')
                flow.train(obj, sample_size, batch_size, epochs, lr, weight_decay, show_progress=False)
                save_object(flow, save_path+name+str(j))
            except ValueError:
                print('Value error')
            except RuntimeError:
                print('Runtime error')
            else:
                break
        print(name + str(j))
```
